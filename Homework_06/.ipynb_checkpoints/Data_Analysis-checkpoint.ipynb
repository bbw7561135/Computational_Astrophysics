```python
import os, re
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
from mpl_toolkits.axes_grid1.inset_locator import inset_axes
from scipy.optimize import curve_fit
```

```python
def LINEAR_FIT(X,a,b):
    return a*X+b
```

### Plot Field Distribution

```python
data_field_analytic = np.loadtxt("./analytical_field_distribution.txt")
data_field_simulatd = np.loadtxt("./simulated_field_distribution_allocate.txt")
# data_rho = np.loadtxt("./charge_distribution.txt")
N = data_field_analytic.shape[0]
L = 1.0
dx = L/N
xx, yy = np.meshgrid(np.arange(N), np.arange(N))
```

```python
# phi_analytic = np.exp(dx*xx-dx*yy)*dx*xx*(1.-dx*xx)*dx*yy*(1.-dx*yy)
# rho = (2*dx*xx*(dx*yy-1.)*(dx*yy-2.*dx*xx+dx*xx*dx*yy+2.)*np.exp(dx*xx-dx*yy))
# phi_analytic = (1.0-yy)
```

```python
data_field_simulatd.shape
```

```python
fig = plt.figure(figsize=(2*6,6), dpi=300)

ax_1 = fig.add_subplot(1, 2, 1, projection='3d')
surf = ax_1.plot_surface(dx*xx, dx*yy, data_field_simulatd, cmap='jet', edgecolor='none', )
ax_1.set_title('Simulated Potential', fontsize=20, y=1.05)
ax_1.view_init(90,-90)
ax_1.axis('tight')
ax_1.set_xlim([0,L])
ax_1.set_ylim([0,L])
ax_1.set_xticks(np.linspace(0,L,5))
ax_1.set_yticks(np.linspace(0,L,5))
ax_1.set_xticklabels(np.linspace(0,L,5), fontsize=10., va='center')
ax_1.set_yticklabels(np.linspace(0,L,5), fontsize=10., va='center', rotation=90)
ax_1.set_zticks([])
ax_1.set_xlabel('x', labelpad=10, fontsize=16)
ax_1.set_ylabel('y', labelpad=10, fontsize=16)
axins_1 = inset_axes(ax_1,
                    width="5%",  # width = 50% of parent_bbox width
                    height="40%",  # height : 5%
                    loc='upper right',
                    bbox_to_anchor=(0.01, -0.2, 1, 1), #(position_x, position_y, bbox_size_x, bbox_size_y)
                    bbox_transform=ax_1.transAxes)
fig.colorbar(surf, shrink=0.3, aspect=10, cax=axins_1)

ax_2 = fig.add_subplot(1, 2, 2, projection='3d')
surf = ax_2.plot_surface(dx*xx, dx*yy, data_field_analytic, cmap='jet', edgecolor='none', )
ax_2.set_title('Analytical Potential', fontsize=20, y=1.05)
ax_2.view_init(90,-90)
ax_2.axis('tight')
ax_2.set_xlim([0,L])
ax_2.set_ylim([0,L])
ax_2.set_xticks(np.linspace(0,L,5))
ax_2.set_yticks(np.linspace(0,L,5))
ax_2.set_xticklabels(np.linspace(0,L,5), fontsize=10., va='center')
ax_2.set_yticklabels(np.linspace(0,L,5), fontsize=10., va='center', rotation=90)
ax_2.set_zticks([])
ax_2.set_xlabel('x', labelpad=10, fontsize=16.)
ax_2.set_ylabel('y', labelpad=10, fontsize=16.)
axins_2 = inset_axes(ax_2,
                    width="5%",  # width = 50% of parent_bbox width
                    height="40%",  # height : 5%
                    loc='upper right',
                    bbox_to_anchor=(0.01, -0.2, 1, 1), #(position_x, position_y, bbox_size_x, bbox_size_y)
                    bbox_transform=ax_2.transAxes)
fig.colorbar(surf, shrink=0.3, aspect=10, cax=axins_2)
plt.savefig("/home/lab/Guan_Ming/Computational_Astrophysics/Homework/HW6/Simulated_v.s._Analytic_Potential_N=512_Criteria=1.00000000e-10.png", bbox_inches='tight', dpi=200, pad_inches=0.05)
plt.show()
```

```python
difference = data_field_simulatd-data_field_analytic
ticks=np.linspace(difference.min(), difference.max(),5)

fig = plt.figure(figsize=(3,3), dpi=200)
ax = plt.imshow(difference[::-1,:], cmap='jet') # flip the phi to consistent with the figures above
plt.title("Difference with Analytic Potential", y=1.05)
plt.xlabel('x')
plt.ylabel('y')
plt.xticks(np.linspace(0,N,5), np.linspace(0,L,5))
plt.yticks(np.linspace(0,N,5), np.flip(np.linspace(0,L,5))) # flip the y axis
fig.colorbar(ax, shrink=0.5, aspect=10, ticks=ticks, format='%.2e')
plt.savefig("/home/lab/Guan_Ming/Computational_Astrophysics/Homework/HW6/Simulated_v.s._Analytic_Potential_Difference_N=512_Criteria=1.00000000e-10.png", bbox_inches='tight', dpi=200, pad_inches=0.05)
plt.show()
```

### Iterations v.s $\omega$

```python
path = './OPTIMAL_OMEGA/'
temp = os.listdir(path)
lattice_range = []
files = []
for file in temp:
    if 'lattice_size' in file:
        lattice_size = eval(file.split('_')[-1].split('.')[0].split('=')[-1])
        lattice_range.append(lattice_size)
        files.append(file)
        
optimal_omega_dict = dict(zip(lattice_range,files))
lattice = sorted(list(optimal_omega_dict.keys()))
```

```python
plt.figure(figsize=(2,2),dpi=320)
for L in lattice[:-1]: # 1024 is not included
    data_omega = np.loadtxt(path+optimal_omega_dict[L])
    omega_range = data_omega[:,0]
    iterations = data_omega[:,1]
    plt.plot(omega_range, np.log10(iterations), label='N = %d'%L)
plt.xlabel('$\omega$', fontsize=6.)
plt.ylabel("$log_{10}$Iterations", fontsize=6.)
plt.xticks(np.arange(0.5, 2.25,0.25), fontsize=6.)
plt.yticks(np.arange(1.,8.,1.), fontsize=6.)
plt.legend(loc='upper right', prop={'size':2.75})
plt.title("Iterations Number v.s. $\omega$", fontsize=9.0, y=1.01)
plt.savefig("/home/lab/Guan_Ming/Computational_Astrophysics/Homework/HW6/Optimal_Omega_v.s._Lattice_Size_Criteria=1.00000000e-10.png", bbox_inches='tight', dpi=320, pad_inches=0.05)
```

### Iterations v.s. Lattice Size

```python
data_performance_N_lattice = np.loadtxt('./CHANGE_LATTICE_SIZE/N_lattice_OpenMP_N_threads=8_allocate.txt')
lattice = data_performance_N_lattice[:,0]
iterations = data_performance_N_lattice[:,1]
x_range = np.linspace(4.,10.,200)
plt.figure(figsize=(2,2),dpi=320)
plt.scatter(np.log2(lattice), np.log2(iterations), s=5.0, c='red')
popt, pcov = curve_fit(LINEAR_FIT, np.log2(lattice), np.log2(iterations))
plt.plot(x_range, LINEAR_FIT(x_range, popt[0], popt[1]), lw=0.5, ls='--', c='k', label='Slope = %.4f'%popt[0])
plt.xlabel('$log_2$ Lattice', fontsize=8.)
plt.ylabel("$log_2$ Iterations", fontsize=8.)
plt.xticks(np.arange(0.,12.,2.), fontsize=8.)
plt.yticks(np.arange(4.,14.,2.), fontsize=8.)
plt.xlim([3.5,10.5])
plt.ylim([5.,13.])
plt.legend(loc='upper left', prop={'size':5.0})
plt.title("Iterations Number v.s. Lattice Size", fontsize=8.0, y=1.01)
plt.savefig("/home/lab/Guan_Ming/Computational_Astrophysics/Homework/HW6/Iterations_v.s._Lattice_Size_Criteria=5.00000000e-10.png", bbox_inches='tight', dpi=320, pad_inches=0.05)
```

### Performance using Desktop

```python
data_performance_N_threads = np.loadtxt('./CHANGE_THREADS/N_threads_OpenMP_omega=1.990000_lattice_size=512_allocate.txt')
threads = data_performance_N_threads[:,0]
prepare_time = data_performance_N_threads[:,-2]
compute_time = data_performance_N_threads[:,-1]
```

```python
plt.figure(figsize=(3*3,3), dpi=300)
plt.subplots_adjust(wspace=0.5)
x_range = np.arange(0,4.,0.05)
plt.subplot(131)
plt.scatter(np.log2(threads), np.log2(prepare_time), c='red', s=8.0)
popt, pcov = curve_fit(LINEAR_FIT, np.log2(threads)[:8], np.log2(prepare_time)[:8])
plt.plot(x_range, LINEAR_FIT(x_range, popt[0], popt[1]), lw=0.5, ls='--', c='red', label="Slope = %.4f"%popt[0])
plt.xticks(np.arange(0.0,4.0,1))
plt.yticks(np.arange(-10.0,-6.0,0.5))
plt.xlim([-0.5,3.5])
plt.ylim([-10.1,-6.5])
plt.xlabel('$log_2$ OpemMP Threads')
plt.ylabel("$log_2$ Time (s)")
plt.title("Preparation Time v.s. Threads",y=1.02, fontsize=10.0)
plt.legend(loc='upper right',prop={'size':7.0})

plt.subplot(132)
plt.scatter(np.log2(threads), np.log2(compute_time), c='blue', s=8.0)
popt, pcov = curve_fit(LINEAR_FIT, np.log2(threads)[:8], np.log2(compute_time)[:8])
plt.plot(x_range, LINEAR_FIT(x_range, popt[0], popt[1]), lw=0.5, ls='--', c='blue', label="Slope = %.4f"%popt[0])
plt.xticks(np.arange(0.0,4.0,1))
plt.yticks(np.arange(-1.0,3.0,0.5))
plt.xlim([-0.5,3.5])
plt.ylim([-1.0,2.5])
plt.xlabel('$log_2$ OpemMP Threads')
plt.ylabel("$log_2$ Time (s)")
plt.title("Computing Time v.s. Threads",y=1.02,fontsize=10.0)
plt.legend(loc='upper right',prop={'size':7.0})

plt.subplot(133)
plt.scatter(threads, (prepare_time+compute_time[0])/(prepare_time+compute_time)/threads, c='lime', s=8.0)
plt.xticks(np.arange(1.0,10.0,2.))
plt.yticks(np.arange(0.3,1.1,0.1))
plt.xlim([0.75,10.25])
plt.ylim([0.35,1.05])
plt.xlabel('OpemMP Threads')
plt.ylabel("Parallel Efficiency")
plt.title("Parallel Efficiency v.s. Threads",y=1.02, fontsize=10.0)
plt.savefig("/home/lab/Guan_Ming/Computational_Astrophysics/Homework/HW6/Performance_N=512_Criteria=1.00000000e-10.png", bbox_inches='tight', dpi=300, pad_inches=0.05)
```

### Performance of Different Lattice Size

```python
path = './CHANGE_THREADS/'
temp = os.listdir(path)
files = []
lattice_range = []
for file in temp:
    if 'lattice_size' in file and 'notebook' not in file:
        lattice_size = eval(file.split('_')[-2].split('=')[-1])
        lattice_range.append(lattice_size)
        files.append(file)
        
performance_dict = dict(zip(lattice_range,files))
lattice = sorted(list(performance_dict.keys()))
```

```python
plt.figure(figsize=(3*4,4), dpi=300)
plt.subplots_adjust(wspace=0.5)
ax_1 = plt.subplot(131)
ax_2 = plt.subplot(132)
ax_3 = plt.subplot(133)
x_range = np.arange(0,4.,0.05)

for L in lattice[3:]:
    data = np.loadtxt(path+performance_dict[L])
    threads = data[:,0]
    prepare_time = data[:,-2]
    compute_time = data[:,-1]
    
    popt, pcov = curve_fit(LINEAR_FIT, np.log2(threads)[:8], np.log2(prepare_time)[:8])
    ax_1.plot(x_range, LINEAR_FIT(x_range, popt[0], popt[1]), lw=0.5, ls='--', c='k')
    ax_1.scatter(np.log2(threads), np.log2(prepare_time), s=20.0, label='N = %d,  slope = %.4f '%(L,popt[0]))
    ax_1.set_xticks(np.arange(0.0,4.0,1))
    ax_1.set_yticks(np.arange(-17.0,-2.0,2.0))
    ax_1.set_xlim([-0.5,3.5])
    ax_1.set_ylim([-16.1,-3.9])
    ax_1.set_xlabel('$log_2$ OpemMP Threads')
    ax_1.set_ylabel("$log_2$ Time (s)")
    ax_1.set_title("Preparation Time v.s. Threads",y=1.02, fontsize=10.0)
    ax_1.legend(loc='lower left',prop={'size':7.0})
    
    popt, pcov = curve_fit(LINEAR_FIT, np.log2(threads)[:4], np.log2(compute_time)[:4])
    ax_2.plot(x_range, LINEAR_FIT(x_range, popt[0], popt[1]), lw=0.5, ls='--', c='blue')
    ax_2.scatter(np.log2(threads), np.log2(compute_time), s=20.0, label='N = %d, Slope = %.4f'%(L,popt[0]))
    ax_2.set_xticks(np.arange(0.0,4.0,1))
    ax_2.set_yticks(np.arange(-16,10.0,2.))
    ax_2.set_xlim([-0.5,3.5])
    ax_2.set_ylim([-16.5,9.5])
    ax_2.set_xlabel('$log_2$ OpemMP Threads')
    ax_2.set_ylabel("$log_2$ Time (s)")
    ax_2.set_title("Computing Time v.s. Threads",y=1.02,fontsize=10.0)
    ax_2.legend(loc='lower left',prop={'size':7.0})

    ax_3.scatter(threads, (prepare_time+compute_time[0])/(prepare_time+compute_time)/threads, s=20.0, label='N = %d'%L)
    ax_3.set_xticks(np.arange(1.0,10.0,2.))
    ax_3.set_yticks(np.arange(-1.0,1.1,0.1))
    ax_3.set_xlim([-0.25,10.25])
    ax_3.set_ylim([-0.05,1.05])
    ax_3.set_xlabel('OpemMP Threads')
    ax_3.set_ylabel("Parallel Efficiency")
    ax_3.set_title("Parallel Efficiency v.s. Threads",y=1.02, fontsize=10.0)
    ax_3.legend(loc='lower left',prop={'size':7.0})
plt.savefig("/home/lab/Guan_Ming/Computational_Astrophysics/Homework/HW6/Performance_Difference_Lattice_Size.png", bbox_inches='tight', dpi=300, pad_inches=0.05)
```
