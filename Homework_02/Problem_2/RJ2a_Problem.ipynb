```python
import os, re
from shutil import copy2
import numpy as np
import matplotlib.pyplot as plt
```

```python
original_path = "/home/lab/gamer/bin/shocktube_MHD/"
save_path = "/home/lab/Guan_Ming/Computational_Astrophysics/Homework/HW2/Problem_2/"
reference = np.loadtxt("/home/lab/Guan_Ming/Computational_Astrophysics/Homework/HW2/Problem_2/RJ2a_reference.txt")
# comparsion = np.loadtxt("/home/lab/gamer/bin/shocktube_MHD/64x64x64_MAXLEVEL=2/Xline_y0.000_z0.000_000020")
data_list = []
max_level_list= []
for obj in os.listdir(original_path):
    if "16x16x16" in obj:
        print(obj)
        max_level=eval(obj[-1])
        max_level_list.append(max_level)
        copy2(original_path+obj+'/'+'Xline_y0.000_z0.000_000020', save_path+'RJ2a_GAMER_MHD_MAXLEVEL=%d.txt'%max_level)
        data_list.append(np.loadtxt(save_path+'RJ2a_GAMER_MHD_MAXLEVEL=%d.txt'%max_level))
data_dict = dict(zip(max_level_list,data_list))
max_level_list.sort()
```

```python
plt.figure(figsize=(4,4),dpi=200)
plt.plot(reference[:,0], reference[:,1], color='k', label='reference', lw=1.0, ls=":")
# plt.scatter(comparsion[:,3], comparsion[:,6], label='64_MAX_LEVEL=2', s=5.0, marker='o')
for i in max_level_list:
    data = data_dict[i]
    plt.scatter(data[:,3], data[:,6], label='GAMER_MAX_LEVEL=%d'%i, s=5.0, marker='o')
plt.title("RJ2a Problem-Density", fontsize=13.0, y=1.03)
plt.xlabel("$x$")
plt.ylabel("$\\rho$")
plt.legend(loc="best", prop={'size':4.0})
plt.savefig("/home/lab/Guan_Ming/Computational_Astrophysics/Homework/HW2/Problem_2/density.png", bbox_inches='tight', dpi=200, pad_inches=0.05)
plt.show()
```
