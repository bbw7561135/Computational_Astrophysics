```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
LF = np.loadtxt("/home/lab/Guan_Ming/Computational_Astrophysics/Homework/HW2/Problem_1-1/SOD_shock_tube_Lax-Friedrichs.txt")
LW = np.loadtxt("/home/lab/Guan_Ming/Computational_Astrophysics/Homework/HW2/Problem_1-1/SOD_shock_tube_Lax-Wendroff.txt")
MH = np.loadtxt("/home/lab/Guan_Ming/Computational_Astrophysics/Homework/HW2/Problem_1-1/SOD_shock_tube_MUSCL-Hancock.txt")
GAMER = np.loadtxt("/home/lab/Guan_Ming/Computational_Astrophysics/Homework/HW2/Problem_1-1/SOD_shock_tube_GAMER.txt")
reference = np.loadtxt("/home/lab/Guan_Ming/Computational_Astrophysics/Homework/HW2/Problem_1-1/strong_shock.txt")
data_list = [LF, LW, MH, GAMER]
scheme_list = ['Lax-Friedrichs', 'Lax-Wendroff', 'MUSCL-Hancock','GAMER']
marker_list = ["o", "*", "h","^"]
```

```python
plt.figure(figsize=(4*2,4), dpi=200)
plt.subplots_adjust(wspace=0.3)
x = LF[:,0]
plt.subplot(121)
for index, data in enumerate(data_list):
    shift = 100*index
    if scheme_list[index]!="GAMER":
        density = data[:,1]
        plt.scatter(x, density+shift, s=1.0, label="%s"%scheme_list[index], marker=marker_list[index])
    else:
        density = data[:,6]
        plt.scatter(data[:,3],density+shift, s=1.0, label="%s"%scheme_list[index], marker=marker_list[index])
    
    if index == 0:
        plt.plot(x,reference[:,1]+shift, label="Reference Solution", color='k', lw=1.0, ls=":")
    else:
        plt.plot(x,reference[:,1]+shift, color='k', lw=1.0, ls=":")
        
plt.xlabel("Position", fontsize=12.5)
plt.ylabel("$\\rho$", fontsize=12.5)
plt.legend(loc="best", prop={"size":6.5})
plt.title("Density of SOD Shock-Tube", fontsize=16, y=1.02)
plt.xlim(0,1.0)
plt.ylim(0., 1600)
plt.subplot(122)
for index, data in enumerate(data_list):
    shift = 100*index
    if scheme_list[index]!="GAMER":
        density = data[:,1]
        plt.scatter(x, density+shift, s=1.0, label="%s"%scheme_list[index], marker=marker_list[index])
    else:
        density = data[:,6]
        plt.scatter(data[:,3],density+shift, s=1.0, label="%s"%scheme_list[index], marker=marker_list[index])
    
    if index == 0:
        plt.plot(x,reference[:,1]+shift, label="Reference Solution", color='k', lw=1.0, ls=":")
    else:
        plt.plot(x,reference[:,1]+shift, color='k', lw=1.0, ls=":")
        
plt.xlabel("Position", fontsize=12.5)
plt.ylabel("$\\rho$", fontsize=12.5)
plt.legend(loc="upper left", prop={"size":6.5})
plt.title("Zoom-in near \nContact Discontinuity", fontsize=16, y=1.02)
plt.xlim(0.7,0.8)
plt.ylim(400., 900.)
plt.savefig("/home/lab/Guan_Ming/Computational_Astrophysics/Homework/HW2/Problem_1-1/density.png", bbox_inches='tight', dpi=200, pad_inches=0.05)
plt.show()
```

```python
plt.figure(figsize=(4*2,4), dpi=200)
plt.subplots_adjust(wspace=0.3)
x = LF[:,0]
ax_1 = plt.subplot(121)
for index, data in enumerate(data_list):
    shift = 0.1*index
    if scheme_list[index]!="GAMER":
        velocity = data[:,2]
        plt.scatter(x, velocity+shift, s=1.0, label="%s"%scheme_list[index], marker=marker_list[index])
    else:
        velocity = data[:,7]/data[:,6]  #Gamer gives momentum instead of velocity
        plt.scatter(data[:,3],velocity+shift, s=1.0, label="%s"%scheme_list[index], marker=marker_list[index])
    if index == 0:
        plt.plot(x,reference[:,2]+shift, label="Reference Solution", color='k', lw=1.0, ls=":")
    else:
        plt.plot(x,reference[:,2]+shift, color='k', lw=1.0, ls=":")
plt.xlabel("Position", fontsize=12.5)
plt.ylabel("$V_x$", fontsize=12.5)
plt.legend(loc="upper left", prop={"size":6.5})
plt.title("Velocity of SOD Shock-Tube", fontsize=16, y=1.02)
plt.xlim(0,1.0)
plt.ylim(-0.1, 1.2)
ax_2 = plt.subplot(122)
for index, data in enumerate(data_list):
    shift = 0.2*index
    if scheme_list[index]!="GAMER":
        velocity = data[:,2]
        plt.scatter(x, velocity+shift, s=5.0, label="%s"%scheme_list[index], marker=marker_list[index])
    else:
        velocity = data[:,7]/data[:,6]  #Gamer gives momentum instead of velocity
        plt.scatter(data[:,3],velocity+shift, s=5.0, label="%s"%scheme_list[index], marker=marker_list[index])
    if index == 0:
        plt.plot(x,reference[:,2]+shift, label="Reference Solution", color='k', lw=1.0, ls=":")
    else:
        plt.plot(x,reference[:,2]+shift, color='k', lw=1.0, ls=":")
        
plt.xlabel("Position", fontsize=12.5)
plt.ylabel("$V_x$", fontsize=12.5)
plt.legend(loc="lower left", prop={"size":6.5})
plt.title("Zoom-in near Shock", fontsize=16, y=1.02)
plt.xlim(0.898,0.902)
plt.ylim(-0.1, 1.5)
plt.savefig("/home/lab/Guan_Ming/Computational_Astrophysics/Homework/HW2/Problem_1-1/velocity.png", bbox_inches='tight', dpi=200, pad_inches=0.05)
plt.show()
```
