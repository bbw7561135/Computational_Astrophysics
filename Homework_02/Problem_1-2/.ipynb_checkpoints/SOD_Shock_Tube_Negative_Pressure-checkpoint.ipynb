```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
MH = np.loadtxt("/home/lab/Guan_Ming/Computational_Astrophysics/Homework/HW2/Problem_1-2/SOD_shock_tube_MUSCL-Hancock.txt")
GAMER = np.loadtxt("/home/lab/Guan_Ming/Computational_Astrophysics/Homework/HW2/Problem_1-2/SOD_shock_tube_GAMER.txt")
GAMER_dual = np.loadtxt("/home/lab/Guan_Ming/Computational_Astrophysics/Homework/HW2/Problem_1-2/SOD_shock_tube_GAMER_dual.txt")
GAMER_input_rho = np.loadtxt("/home/lab/Guan_Ming/Computational_Astrophysics/Homework/HW2/Problem_1-2/SOD_shock_tube_GAMER_input_rho.txt")
data_list = [GAMER, GAMER_dual,GAMER_input_rho]
scheme_list = ['GAMER','GAMER_Dual_Energy','GAMER_Input_Rho']
marker_list = ["h","o", "*"]
```

```python
plt.figure(figsize=(3*2,3), dpi=200)
plt.subplots_adjust(wspace=0.3)
ax_1 = plt.subplot(121)
for index, data in enumerate(data_list):
    shift = index*0.05
#     if scheme_list[index] == "MH":
#         density = data[:,1]
#         plt.scatter(data[:,0],density+shift, s=1.0, label="%s"%scheme_list[index], marker=marker_list[index])
#     else:
#         #shift = 100*index
#         density = data[:,6]
#         plt.scatter(data[:,3],density+shift, s=1.0, label="%s"%scheme_list[index], marker=marker_list[index])
    density = data[:,6]
    plt.scatter(data[:,3],density+shift, s=1.0, label="%s"%scheme_list[index], marker=marker_list[index])
        
plt.xlabel("Position", fontsize=12.5)
plt.ylabel("$\\rho$", fontsize=12.5)
plt.legend(loc="best", prop={"size":5.})
plt.title("Density of SOD Shock-Tube", fontsize=12, y=1.02)
plt.xlim(0,1.0)
ax_2 = plt.subplot(122)
for index, data in enumerate(data_list):
    shift = index*0.05
#     if scheme_list[index] == "MH":
#         density = data[:,1]
#         plt.scatter(data[:,0],density+shift, s=1.0, label="%s"%scheme_list[index], marker=marker_list[index])
#     else:
#         #shift = 100*index
#         density = data[:,6]
#         plt.scatter(data[:,3],density+shift, s=1.0, label="%s"%scheme_list[index], marker=marker_list[index])
    density = data[:,6]
    plt.scatter(data[:,3],density+shift, s=1.0, label="%s"%scheme_list[index], marker=marker_list[index])
plt.xlabel("Position", fontsize=12.5)
plt.ylabel("$\\rho$", fontsize=12.5)
plt.legend(loc="best", prop={"size":5.})
plt.title("Zoom-in near Shock", fontsize=12, y=1.02)
plt.xlim(0.72,0.85)
plt.savefig("/home/lab/Guan_Ming/Computational_Astrophysics/Homework/HW2/Problem_1-2/density.png", bbox_inches='tight', dpi=200, pad_inches=0.05)
plt.show()
```
